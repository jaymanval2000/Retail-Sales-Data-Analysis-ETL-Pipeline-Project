## Step 5- Splitting Dependent & Independent Variable(X & Y)


```python
df.columns
```

```python
y = df['Sales']           # Dependent Variable(y)  # Target Column
```

```python
x = df[['Price  ', 'Quantity']]   # Independent Variable(x)
```

## Step 6- Setting the data for Training and Testing Split


```python
pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=101)
```

## Step 7- Loading the Algorithm

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
prediction = model.predict(X_test)
```

## Step 8- Predicting the results

##  Create a scatterplot of the real test values versue the predicted values

```python
plt.scatter(y_test,prediction)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test,prediction))
print('MSE:', metrics.mean_squared_error(y_test,prediction))
```
